```
#Define function to cleaning and make data in title format.

def cleaning(data):
  clean = {'-','_'}
  
  for char in clean:
    data = data.str.replace(char, ' ')
  return data

def title(data):
  return data.str.title()
```

```
#Clean ‘-‘,’_’ character, and make them in title format (first character capitalized) and all uppercase for VISA

#Restoaccepts : cleaning and title format

ra.Rpayment = cleaning(ra.Rpayment)
ra.Rpayment = title(ra.Rpayment)
ra.Rpayment = ra.Rpayment.replace('Visa','VISA')

#Userpayment : cleaning and title format

up.Upayment = cleaning(up.Upayment)
up.Upayment = title(up.Upayment)
up.Upayment = up.Upayment.replace('Visa','VISA')

#Restocuisine : cleaning

rc.Rcuisine = cleaning(rc.Rcuisine)

#Usercuisine : cleaning

uc.Rcuisine = cleaning(uc.Rcuisine)
```

```
#Transform table, drop duplicates, and change NaN value to 999

rh = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Python For Data Science Project/Week 4_Final Project fix/data/restohours.csv')
rh
```

```
rh_copy = rh.copy()

rh_copy = rh_copy.drop_duplicates()
rh_copy = rh_copy.assign(days=rh_copy.days.str.split(';')).explode('days')

rh_copy = rh_copy[rh_copy.days!='']

rh_copy = rh_copy.groupby(['placeID', 'days'])['hours'].aggregate('first').unstack(fill_value=9999)
rh_copy = rh_copy[['Mon','Tue','Wed','Thu','Fri','Sat','Sun']]
rh_copy
```

```
#Student rate the most

#Read csv.
rating_final = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Python For Data Science Project/Week 4_Final Project fix/data/rating_final.csv')
user_profile = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Python For Data Science Project/Week 4_Final Project fix/data/userprofile.csv')

#Filtering so user_activity dataframe only contains userID and activity column and also user with student activity.
user_activity = user_profile[['userID', 'activity']][user_profile.activity == 'student']

#Joining rating_final to user_activity to make student_rating dataframe.
student_rating = user_activity.join(rating_final.set_index('userID'), on='userID' ,how='inner')

#Getting most rated restaurant(placeID) by counting every student that rate it.
sr_most = student_rating.placeID.value_counts()

#Make sr_most series into new dataframe.
df_sr_most = pd.DataFrame({'placeID':sr_most.index, 'total_rate':sr_most.values})
df_sr_most

#Print the result.
print("So three most rated restaurant ID by student is: {0}, {1}, {2}\n\n {3}" .format(sr_most.index[0], sr_most.index[1], sr_most.index[2], df_sr_most.head(3))) 
```

```
#Five most visited restaurant
#With the assumption that every userID count means visit, then most userID count = the most visited.

most_visited = rating_final.groupby('placeID')['userID'].count().sort_values(ascending=False)
most_visited = pd.DataFrame({'placeID':most_visited.index, 'total_user_visit':most_visited.values})
most_visited
```

```
#Graph
```

```
#Calculating the highest rating to prove wheter the most visited restaurant mean having the highest rating

ratings = ['rating', 'food_rating', 'service_rating']
resto_rating = rating_final.groupby('placeID')[ratings].sum()

resto_rating['total_rating'] = resto_rating.rating + resto_rating.food_rating + resto_rating.service_rating

highest_rating = resto_rating.groupby('placeID')['total_rating'].sum().sort_values(ascending=False)

highest_rating = pd.DataFrame({'placeID':highest_rating.index, 'total_rating':highest_rating.values})
highest_rating
```

```
#Prove wheter the most visited restaurant mean having the highest rating

most_visited.loc[0]['placeID'] == highest_rating.loc[0]['placeID']
```

```
#Five restaurant having highest avarage rating
#avg rating = sum(𝑟𝑎𝑡𝑖𝑛𝑔, 𝑓𝑜𝑜𝑑 𝑟𝑎𝑡𝑖𝑛𝑔, 𝑠𝑒𝑟𝑣𝑖𝑐𝑒 𝑟𝑎𝑡𝑖𝑛𝑔)/3 /Num.of restaurant rating

resto_rating['avg_total_rating'] = resto_rating.total_rating / 3
resto_rating['num_rating'] = rating_final.placeID.value_counts()
resto_rating['avg_rating'] = resto_rating.avg_total_rating / resto_rating.num_rating

resto_rating = resto_rating[['rating', 'food_rating', 'service_rating', 'avg_rating']]
resto_rating.sort_values(by='avg_rating', ascending=False).head(5)
```

```
#Joining avg_rate to df dataframe so there will be a avg_rating column

avg_rate = resto_rating[['avg_rating']]
df = df.join(avg_rate, on='placeID', how='inner')
df.head()
```

```
#Recommend a restaurant
#Japanese restaurant, around city of San Luis Potosi, cashless payment, highest avg rating.

recommended = df.loc[(df.Rcuisine == 'Japanese') & (df.city == 'San Luis Potosi') & (df.Rpayment != 'Cash')]
recommended = recommended.sort_values(by='avg_rating', ascending=False)
print("This is recommended restaurant based on your requirements: \n\n{0}" .format(recommended.iloc[0]))
```
